Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House():
  def __init__(self, area, n_rooms, n_bathrooms, n_floors, garage=False, garden=False, price_sqm=1000):
    self.area = area
    self.n_rooms = n_rooms
    self.n_bathrooms = n_bathrooms
    self.n_floors = n_floors
    self.garage = garage
    self.garden = garden
    self.price_sqm = price_sqm

  def price_calculation(self):
    house_prise = self.area*self.price_sqm

    if self.garage:
      house_prise += 1500
    if self.garden:
      house_prise += 2000
    return house_prise

  def __str__(self):
    return f"The price of the house {self.price_calculation()} EUR"



```

```python
house_1 = House(200,6,3,2,garage=False,garden=True,price_sqm=1200)
print(house_1)
```

Create a Person class with the following attributes:


*   name
*   age
*   married (T/F)
*   number of children
*   number of cars
*   total amount of money (to buy a house)
*   based on the number of children store the number of family members in an attribute  (- this does not have to be included in the init arguments as it is a calculated value)


(4 points)




```python
class Person():
  def __init__(self,name,age,money,married=False,n_children=0,n_cars=0):
    self.name = name
    self.age = age
    self.money = money
    self.married = married
    self.n_children = n_children
    self.n_cars = n_cars

    if self.married == True:
      members = 2
    else:
      members = 1

    self.total_members = members + self.n_children

  def __str__(self):
    return f"In {self.name}'s family {self.total_members} members"

```

```python
person_1 = Person("John", 35, 100000, married=True, n_children=1, n_cars=1)
print(person_1)
```

Create functions (decide whether it should be an instance method of one of the classes, or should be just defined in the program) to answer


*   whether the person can buy the house,
*   whether the person's family can live in that house (enough room),
*   whether the house has enough space (garages) for the cars in the family

(5 points)


```python
class Customer(Person):
    def can_buy(self, house):
        return self.money >= house.price_calculation()

    def has_enough_rooms(self, house):
        return house.n_rooms >= self.total_members

    def has_enough_garage(self, house):
        return not (self.n_cars > 0 and not house.garage)
```
